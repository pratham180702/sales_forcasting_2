```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('Property_Sales_History_Property_Sales_History.csv')
```

```python
df.head()
df.tail()
```

```python
#the SALEPRICE here is in string and has "$" prefix. so first replacing "$" with empty space
df['SALEPRICE'] = df['SALEPRICE'].str.replace('$', '')

#now typecasting the string to float
df['SALEPRICE'] = df['SALEPRICE'].astype(float)
```

```python
#filling the null values using the interpolate method(try using other methods like linear, quadratic instead of cubic)
df['SALEPRICE'].interpolate(method='cubic', inplace=True)
df2 = df[['SALEDATE','SALEPRICE']]
df2
```

```python
#grouping the data based upon dates and adding all the SALEPRICE of the single date
grouped_df = df.groupby(['SALEDATE'])['SALEPRICE'].sum()
grouped_df
```

```python
#resetting the index
grouped_df = grouped_df.reset_index()
grouped_df.head(100)
```

```python
#looking out for maximum SALEPRICE value present in the dataframe 
max_value = df['SALEPRICE'].max()
print(max_value)
```

```python
#to enhanse the quality of the graph
%config InlineBackend.figure_format = 'retina'

#max and min value range of y axis
plt.ylim(0.00, max_value)


#plotting of the graph
plt.plot(grouped_df.index,grouped_df['SALEPRICE'])
```

```python
grouped_df.head(50)
```

```python
# Convert the 'date' column to a datetime
grouped_df['SALEDATE']=pd.to_datetime(grouped_df['SALEDATE'])
# Set the 'date' column as the index of the DataFrame
grouped_df=grouped_df.set_index('SALEDATE')
# Set the 'date' column as the index of the DataFrame
resample_df=grouped_df.resample('Y').mean()
# Group the data by the year
grouped_df1=resample_df.groupby(resample_df.index.year).mean()
```

```python
grouped_df1=grouped_df1.reset_index()
```

```python
grouped_df1
```

```python
df_final = grouped_df1
grouped_df1
```

```python
df_final.drop('index',axis=1,inplace=True)
```

```python
# df_final.drop('SALEDATE',axis=1,inplace=True)

df_final
```

```python
df_final.plot(x='SALEYEAR', y='SALEPRICE', kind='bar')

plt.show()
```

```python

from statsmodels.tsa.arima.model import ARIMA

# Set the values of p, d, and q
p = 1
d = 0
q = 0

# Fit the model
model = ARIMA(df_final['SALEPRICE'], order=(p, d, q))
model_fit = model.fit()

# Make predictions
predictions = model_fit.forecast(steps=1)
```

```python
model_fit.summary()
```

```python
# Plot the original data and predictions
from matplotlib import pyplot as plt
plt.plot(df_final['SALEPRICE'], label='Original Data')
plt.plot(predictions, label='Predictions')
plt.legend()
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(df_final[['SALEYEAR']], df_final['SALEPRICE'], test_size=0.2)

# Instantiate the model
rf = RandomForestRegressor()

# Fit the model on the training data
rf.fit(X_train, y_train)

# Make predictions on the test data
y_pred = rf.predict(X_test)

# Evaluate the model's performance
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, y_pred)
print("Mean Absolute Error:", mae)
```

# Linear regression 

```python
from sklearn.linear_model import LinearRegression

# create a Linear Regression model object
reg = LinearRegression()

# fit the model to the data
reg.fit(df_final[['SALEYEAR']], df_final['SALEPRICE'])

# predict next year's sales
next_year_sale_prediction = reg.predict([[2017],[2018]])

# print the prediction
print(next_year_sale_prediction)
```

```python
import matplotlib.pyplot as plt

# create a scatter plot of the data
plt.bar(df_final['SALEYEAR'], df_final['SALEPRICE'])

# plot the regression line
plt.plot(df_final['SALEYEAR'], reg.predict(df_final[['SALEYEAR']]), color='red')

# create a bar plot of the data
plt.bar(2017, next_year_sale_prediction[0])
plt.bar(2018, next_year_sale_prediction[1])

# add labels and title
plt.xlabel('Year')
plt.ylabel('Sale Price')
plt.title('Predicted Sale Price for 2022')

plt.annotate(f'Predicted Sale Price for 2018: ${next_year_sale_prediction[1]:.2f}',
             xy=(2018, next_year_sale_prediction[1]),
             xytext=(2018, next_year_sale_prediction[1] + 1000),
             arrowprops=dict(facecolor='red', shrink=0.05))
# show the plot
plt.show()


```
